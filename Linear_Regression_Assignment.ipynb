#Commodity Price Prediction

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_csv('/content/commodity_price.csv')
print(df.head())
```

```python
df.dropna()
```

```python
df_encoded = pd.get_dummies(df, columns=[ 'State', 'District', 'Market', 'Commodity', 'Variety' ], drop_first= True)
```

```python
X = df_encoded[['Min_x0020_Price', 'Max_x0020_Price'] + [col for col in df_encoded.columns if col.startswith(('State_', 'District_', 'Market_', 'Commodity_', 'Variety_', 'Grade_'))]]
y = df_encoded['Modal_x0020_Price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
model= LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f'Mean Squared Eroor : {mse}')
print(f'r2 Score : {r2}')
```
