### Text Pre-processing

This notebook demonstrates various text pre-processing techniques on two sample paragraphs:

1.  **Stop word removal**: Eliminating common words that don't add much meaning (e.g., "the", "is", "a").
2.  **Removing punctuation marks**: Getting rid of characters like periods, commas, question marks, etc.
3.  **Stemming**: Reducing words to their root or base form (e.g., "running" -> "run").
4.  **Lemmatization**: Reducing words to their dictionary or lexical base form, considering vocabulary and morphological analysis (e.g., "better" -> "good").

```python
# Define the two input paragraphs
paragraphs = (
    "NLTK (Natural Language Toolkit) is a powerful library for working with human language data in Python. It provides easy-to-use interfaces to over 50 corpora and lexical resources such as WordNet, along with a suite of text processing libraries for classification, tokenization, stemming, tagging, parsing, and semantic reasoning.",
    "Text pre-processing is a crucial step in Natural Language Processing (NLP) tasks, often involving techniques like tokenization, stop word removal, and lemmatization. These steps help clean and standardize text data, which is essential for accurate analysis and model training! Are you ready to learn?"
)

# Combine into a single input string
input_text = " ".join(paragraphs)
print("Original Text:")
print(input_text)
```

### 1. Import Libraries and Download NLTK Data

I'll use NLTK for tokenization, stop word removal, stemming, and lemmatization. We need to download specific data packages first.

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer
import string

# Download necessary NLTK data directly
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('punkt_tab') # Explicitly download punkt_tab as suggested by error message
nltk.download('wordnet')
nltk.download('omw-1.4') # Often needed for WordNetLemmatizer
```

### 2. Tokenization

First, we tokenize the text into words, which is a prerequisite for most other pre-processing steps.

```python
# Tokenize the text
words = word_tokenize(input_text.lower()) # Convert to lowercase for consistent processing
print("\nTokens after lowercasing:")
print(words[:20], "...") # Display first 20 tokens
```

### 3. Removing Punctuation Marks

Next, we remove punctuation from the tokens.

```python
# Remove punctuation
punctuation_free_words = [word for word in words if word.isalnum()]
print("\nTokens after punctuation removal:")
print(punctuation_free_words[:20], "...")
```

### 4. Stop Word Removal

Now, we filter out common stop words.

```python
# Get English stop words
stop_words = set(stopwords.words('english'))

# Remove stop words
filtered_words = [word for word in punctuation_free_words if word not in stop_words]
print("\nTokens after stop word removal:")
print(filtered_words[:20], "...")
```

### 5. Stemming

Applying stemming to reduce words to their root form.

```python
# Initialize Porter Stemmer
porter = PorterStemmer()

# Perform stemming
stemmed_words = [porter.stem(word) for word in filtered_words]
print("\nTokens after stemming:")
print(stemmed_words[:20], "...")
```

### 6. Lemmatization

Finally, we apply lemmatization for a more linguistically informed reduction to the base form.

```python
# Initialize WordNet Lemmatizer
lemmatizer = WordNetLemmatizer()

# Perform lemmatization
lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_words]
print("\nTokens after lemmatization:")
print(lemmatized_words[:20], "...")
```
